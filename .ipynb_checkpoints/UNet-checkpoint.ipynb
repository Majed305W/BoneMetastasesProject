## Dataloader

```python
import os
import torch
import pathlib
import imageio.v2 as imageio  # Ensure proper image reading
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import numpy as np
import torchvision.transforms as transforms

class BoneMets(Dataset):
    def __init__(self, images_dir: str, masks_dir: str, transforms=None):
        self.images_list = sorted(list(pathlib.Path(images_dir).glob('*.png')))
        self.masks_list = sorted(list(pathlib.Path(masks_dir).glob('*.png')))
        assert len(self.images_list) == len(self.masks_list), "Mismatch between images and masks count"
        self.transforms = transforms

    def __len__(self):
        return len(self.images_list)
    def __getitem__(self, idx):
        image = imageio.imread(self.images_list[idx])  # (H, W, 3) for RGB or (H, W) for grayscale
        mask = imageio.imread(self.masks_list[idx])  # (H, W), assumes single-channel
        image = Image.fromarray(image)  
        mask = (mask > 0).astype(np.uint8)
        mask = torch.from_numpy(mask).unsqueeze(0).float()  
        if self.transforms:
            image = self.transforms(image)  
        return image, mask

transform = transforms.Compose([
    transforms.ToTensor() ])
    
# Define dataset directories
train_images_dir = r'/mnt/gpussd2/pawansj/Pawan/Files/3Torch/2024-T/5BM_Seg_Data/Data/Split/Train/image1'
train_masks_dir = r'/mnt/gpussd2/pawansj/Pawan/Files/3Torch/2024-T/5BM_Seg_Data/Data/Split/Train/mask1'
valid_images_dir = r'/mnt/gpussd2/pawansj/Pawan/Files/3Torch/2024-T/5BM_Seg_Data/Data/Split/Valid/image1'
valid_masks_dir = r'/mnt/gpussd2/pawansj/Pawan/Files/3Torch/2024-T/5BM_Seg_Data/Data/Split/Valid/mask1'
test_images_dir = r'/mnt/gpussd2/pawansj/Pawan/Files/3Torch/2024-T/5BM_Seg_Data/Data/Split/Test/image1'
test_masks_dir = r'/mnt/gpussd2/pawansj/Pawan/Files/3Torch/2024-T/5BM_Seg_Data/Data/Split/Test/mask1'

# Load datasets
train_data = BoneMets(images_dir=train_images_dir, masks_dir=train_masks_dir, transforms=transform)
valid_data = BoneMets(images_dir=valid_images_dir, masks_dir=valid_masks_dir, transforms=transform)
test_data = BoneMets(images_dir=test_images_dir, masks_dir=test_masks_dir, transforms=transform)

# Sanity check
image, mask = train_data[0]
print(f'Train Image shape: {image.shape}')  # Expected: (C, H, W)
print(f'Train Mask shape: {mask.shape}')  # Expected: (1, H, W) for binary mask

# Check unique values in mask
print(f'Unique values in mask: {torch.unique(mask)}')  # Should print tensor([0., 1.])

# Create DataLoaders
train_dataloader = DataLoader(train_data, batch_size=8, num_workers=os.cpu_count(), shuffle=True)
valid_dataloader = DataLoader(valid_data, batch_size=8, num_workers=os.cpu_count(), shuffle=False)
test_dataloader = DataLoader(test_data, batch_size=8, num_workers=os.cpu_count(), shuffle=False)

# Print dataset sizes
print(f'Number of images in the train set: {len(train_data)}')
print(f'Number of images in the valid set: {len(valid_data)}')
print(f'Number of images in the test set: {len(test_data)}')
```

## Model

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class DoubleConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(DoubleConv, self).__init__()
        self.conv_block = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1, stride=1, bias=False),
            nn.BatchNorm2d(num_features=out_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, stride=1, bias=False),
            nn.BatchNorm2d(num_features=out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.conv_block(x)

class Unet(nn.Module):
    def __init__(self, in_channels, out_channels, features=[32, 64, 128, 256, 512]):
        super(Unet, self).__init__()
        self.downs = nn.ModuleList()
        self.ups = nn.ModuleList()
        self.max_pool = nn.MaxPool2d(kernel_size=2)

        # Define encoder (downsampling path)
        for feature in features:
            self.downs.append(DoubleConv(in_channels, feature))
            in_channels = feature

        # Define bottleneck and decoder (upsampling path)
        self.bottle_neck = DoubleConv(features[-1], features[-1] * 2)
        self.out_conv = nn.Conv2d(features[0], out_channels, kernel_size=1)

        for feature in reversed(features):
            self.ups.append(nn.ConvTranspose2d(feature * 2, feature, kernel_size=2, stride=2))
            self.ups.append(DoubleConv(feature * 2, feature))

    def forward(self, x):
        skip_connections = []

        # Encoder (downsampling path)
        for down in self.downs:
            x = down(x)
            skip_connections.append(x)
            x = self.max_pool(x)

        x = self.bottle_neck(x)

        # Decoder (upsampling path)
        for idx in range(0, len(self.ups), 2):
            x = self.ups[idx](x)
            skip_connection = skip_connections.pop()
            if x.shape != skip_connection.shape:
                x = F.interpolate(x, size=skip_connection.shape[2:], mode='bilinear', align_corners=True)
            x = torch.cat((skip_connection, x), dim=1)
            x = self.ups[idx + 1](x)

        return self.out_conv(x)

# Example usage
model = Unet(in_channels=1, out_channels=1)
output = model(torch.randn(1, 1, 512, 512))
print(output.shape)

summary(model,input_size=(1,1,512,512),col_width=20,col_names=['input_size','output_size','num_params','trainable'],row_settings=['var_names'])
```

```python
loss_fn = torch.nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
def train_step(model: torch.nn.Module,
               train_dataloader: torch.utils.data.DataLoader,
               valid_dataloader: torch.utils.data.DataLoader,
               loss_fn: torch.nn.Module,
               optimizer: torch.optim.Optimizer,
               device: torch.device = torch.device('cpu')):
    
    train_loss = 0
    num_batches = len(train_dataloader)
    model.to(device)
    model.train()
    
    for batch, (X_train, y_train) in enumerate(train_dataloader):
        X_train = X_train.to(device)
        y_train = y_train.to(device)
        optimizer.zero_grad()
        
        y_logits = model(X_train)
        loss = loss_fn(y_logits, y_train)
        train_loss += loss.item()
        
        loss.backward()
        optimizer.step()
    
    train_loss /= len(train_dataloader)
    
    val_loss = validate(model, valid_dataloader, loss_fn, device)
    
    return train_loss, val_loss

def validate(model: torch.nn.Module,
             valid_dataloader: torch.utils.data.DataLoader,
             loss_fn: torch.nn.Module,
             device: torch.device = torch.device('cpu')):
    
    model.eval()
    val_loss = 0
    
    with torch.no_grad():
        for X_val, y_val in valid_dataloader:
            X_val = X_val.to(device)
            y_val = y_val.to(device)
            
            y_val_logits = model(X_val)
            val_loss += loss_fn(y_val_logits, y_val).item()
    
    val_loss /= len(valid_dataloader)
    
    return val_loss
```

```python
import torch
from tqdm.auto import tqdm

epochs = 50

train_losses = []
val_losses = []

for epoch in tqdm(range(epochs)):
    train_loss, val_loss = train_step(model=model, train_dataloader=train_dataloader, valid_dataloader=valid_dataloader, loss_fn=loss_fn, optimizer=optimizer, device=device)
    
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    
    print(f'Epoch {epoch + 1}: Train Loss: {train_loss}, Val Loss: {val_loss}')
```

```python
import torch
import numpy as np

def dice_score(pred, target, threshold=0.5):
    pred = (pred > threshold).float()
    target = target.float()
    intersection = (pred * target).sum()
    return (2. * intersection) / (pred.sum() + target.sum() + 1e-8)

def precision_score(pred, target, threshold=0.5):
    pred = (pred > threshold).float()
    target = target.float()
    tp = (pred * target).sum()
    fp = ((pred == 1) & (target == 0)).sum()
    return tp / (tp + fp + 1e-8)

def recall_score(pred, target, threshold=0.5):
    pred = (pred > threshold).float()
    target = target.float()
    tp = (pred * target).sum()
    fn = ((pred == 0) & (target == 1)).sum()
    return tp / (tp + fn + 1e-8)
```
